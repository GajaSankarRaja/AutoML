```python
import pandas as pd
import numpy as np
d=pd.read_csv('/content/drive/MyDrive/Datasets/dataorg.csv')
d=d[['RICE PRODUCTION (1000 tons)','RICE YIELD (Kg per ha)','RICE AREA (1000 ha)']]
```

```python
#CHECKING FOR THE CORRELATION:
import seaborn as sns
sns.heatmap(d.corr(),annot=True)
```

```python
#SPLLITTING THE FEATURES AND TARGET:
x=d[['RICE PRODUCTION (1000 tons)','RICE AREA (1000 ha)']]
y=d['RICE YIELD (Kg per ha)']
```

```python
#TRAINNING THE DATASET:
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#FITTING THE REGRESSION MODEL1:
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
rg.fit(x_train,y_train)
```

```python
#CHECKING THE MODEL SCORE AND ERRORS:
ypred1=rg.predict(x_test)
from sklearn.metrics import r2_score,mean_squared_error
print("R2 SCORE:",r2_score(y_test,ypred1))
print("MSE",mean_squared_error(y_test,ypred1))
```
